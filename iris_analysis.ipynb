## 1. Load iris.csv (use numpy, NO pandas) into numpy array called “iris”

```python
#import numpy & matplotlib but not pandas
import numpy as np
import matplotlib.pyplot as plt

# Import the iris data from the CSV file as strings
sepal_length, sepal_width, petal_length, petal_width, classes = np.loadtxt('iris.csv', delimiter=',', skiprows=1,  dtype=str, unpack=True)
# Convert to floating-point values except classes
sepal_length = sepal_length.astype(float)
sepal_width = sepal_width.astype(float)
petal_length = petal_length.astype(float)
petal_width = petal_width.astype(float)
```

## 2. Analyze - (1) Plot

```python
# 2. Analyze - (1) Plot 

# create 6 subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# (1) Scatter: Sepal Length vs Sepal Width (red)
axs[0,0].scatter(sepal_length, sepal_width, c='r', alpha=0.7)
axs[0,0].set_title('Sepal Length vs Width')
axs[0,0].set_xlabel('Sepal Length')
axs[0,0].set_ylabel('Sepal Width')

# (2) Petal Length vs Petal Width (green)
axs[0,1].scatter(petal_length, petal_width, c='g', alpha=0.7)
axs[0,1].set_title('Petal Length vs Width')
axs[0,1].set_xlabel('Petal Length')
axs[0,1].set_ylabel('Petal Width')

# (3) Scatter: Sepal Length vs Petal Length (orange)
axs[1,0].scatter(sepal_length, petal_length, c='orange',alpha=0.7)
axs[1,0].set_title('Sepal Length vs Petal Length')
axs[1,0].set_xlabel('Sepal Length')
axs[1,0].set_ylabel('Petal Length')

# (4) Scatter: Sepal Width vs Petal Width (skyblue)
axs[1,1].scatter(sepal_width, petal_width, c='skyblue',alpha=0.7)
axs[1,1].set_title('Sepal Width vs Petal Width')
axs[1,1].set_xlabel('Sepal Width')
axs[1,1].set_ylabel('Petal Width')

plt.tight_layout()  # prevent overlap
plt.show() #render plots
```

## 2. Analyze - (2) simple statements 

**1. Sepal Length vs. Sepal Width**

Variation in sepal length doesn't really show a clear link to sepal width for all the species.
The sepal length doesn't give a reliable clue about the sepal width.

**2. Petal Length vs. Petal Width**

Petal length & width have a strong *linear* correlation. The petal width tends to grow as the petal length increases. 

**3.Sepal Length vs. Petal Length**

There's a somewhat *linear* correlation, but it's not as evident as between petal length and petal width.

**4. Sepal Width vs Petal Width**

There isn't a distinct correlation between Sepal Width and Petal Width.



## 2. Analyze - (3) Calculations/stats (minimum, maximum, average, ...)


```python
# 2. Analyze - (3) Calculations/stats (minimum, maximum, average, ...)
'''
display statistics (mean & standard deviation)
    
Args:
- attribute (numpy array) : the attribute data
- name (str) : the name of the attribute data
'''
def display_stats(attribute, name):
    print(name)
    print("\t mean:", np.mean(attribute), "\t max:", np.max(attribute), "\t min:", np.min(attribute), "\t std dev:", np.std(attribute))

# Calculations/stats (minimum, maximum, average, ...)
display_stats(sepal_length, "Sepal Length")
display_stats(sepal_width, "Sepal Width")
display_stats(petal_length, "Petal Length")
display_stats(petal_width, "Petal Width")
```

## 3. Create separate numpy arrays for the three classes called: “setosa”, “versicolor” and “virginica”


```python
# (1) Setosa
setosa_sepal_length = sepal_length[classes == "setosa"]
setosa_sepal_width = sepal_width[classes == "setosa"]
setosa_petal_length = petal_length[classes == "setosa"]
setosa_petal_width = petal_width[classes == "setosa"]

# (2) Versicolor
versicolor_sepal_length = sepal_length[classes == "versicolor"]
versicolor_sepal_width = sepal_width[classes == "versicolor"]
versicolor_petal_length = petal_length[classes == "versicolor"]
versicolor_petal_width = petal_width[classes == "versicolor"]

# (3) Virginica
virginica_sepal_length = sepal_length[classes == "virginica"]
virginica_sepal_width = sepal_width[classes == "virginica"]
virginica_petal_length = petal_length[classes == "virginica"]
virginica_petal_width = petal_width[classes == "virginica"]
```

## 4. Compare the different classes (use matplotlib to visualize)

```python
# 4. Compare the different classes (use matplotlib to visualize)
fig, axs = plt.subplots(3, 2, figsize=(12, 8))

# (1) Sepal Length vs Sepal Width
axs[0, 0].scatter(setosa_sepal_length, setosa_sepal_width, label="Setosa",color='r', alpha=0.7)
axs[0, 0].scatter(versicolor_sepal_length, versicolor_sepal_width, label="Versicolor",color='g', alpha=0.7)
axs[0, 0].scatter(virginica_sepal_length, virginica_sepal_width, label="Virginica",color='b', alpha=0.7)
axs[0, 0].set_title('Sepal Length vs Width')
axs[0, 0].set_xlabel('Sepal Length')
axs[0, 0].set_ylabel('Sepal Width')
axs[0, 0].legend()

# (2) Petal Length vs Petal Width
axs[0, 1].scatter(setosa_petal_length, setosa_petal_width, label="Setosa",color='r', alpha=0.7)
axs[0, 1].scatter(versicolor_petal_length, versicolor_petal_width, label="Versicolor",color='g', alpha=0.7)
axs[0, 1].scatter(virginica_petal_length, virginica_petal_width, label="Virginica", color='b',alpha=0.7)
axs[0, 1].set_title('Petal Length vs Width')
axs[0, 1].set_xlabel('Petal Length')
axs[0, 1].set_ylabel('Petal Width')
axs[0, 1].legend()

# (3) Sepal Length vs Petal Length
axs[1, 0].scatter(setosa_sepal_length, setosa_petal_length, label="Setosa",color='r', alpha=0.7)
axs[1, 0].scatter(versicolor_sepal_length, versicolor_petal_length, label="Versicolor",color='g', alpha=0.7)
axs[1, 0].scatter(virginica_sepal_length, virginica_petal_length, label="Virginica", color='b',alpha=0.7)
axs[1, 0].set_title('Sepal Length vs Petal Length')
axs[1, 0].set_xlabel('Sepal Length')
axs[1, 0].set_ylabel('Petal Length')
axs[1, 0].legend()

# (4) Scatter: Sepal Width vs Petal Width
axs[1, 1].scatter(setosa_sepal_width, setosa_petal_width, label="Setosa",color='r', alpha=0.7)
axs[1, 1].scatter(versicolor_sepal_width, versicolor_petal_width, label="Versicolor",color='g', alpha=0.7)
axs[1, 1].scatter(virginica_sepal_width, virginica_petal_width, label="Virginica", color='b',alpha=0.7)
axs[1, 1].set_title('Sepal Width vs Petal Width')
axs[1, 1].set_xlabel('Sepal Width')
axs[1, 1].set_ylabel('Petal Width')
axs[1, 1].legend()

# (5) Histogram: Sepal Length across species
axs[2, 0].hist(setosa_sepal_length, bins=10, alpha=0.7, rwidth=0.85,color='r',label='Setosa')
axs[2, 0].hist(versicolor_sepal_length, bins=10, alpha=0.7, rwidth=0.85, color='g',label='Versicolor')
axs[2, 0].hist(virginica_sepal_length, bins=10, alpha=0.7, rwidth=0.85, color='b',label='Virginica')
axs[2, 0].set_title('Sepal Length Distribution across Species')
axs[2, 0].set_xlabel('Sepal Length')
axs[2, 0].set_ylabel('Frequency')
axs[2, 0].legend()

# (6) Histogram : Petal Length across species
axs[2, 1].hist(setosa_petal_length, bins=10, alpha=0.7,  rwidth=0.85, color='r',label='Setosa')
axs[2, 1].hist(versicolor_petal_length, bins=10, alpha=0.7, rwidth=0.85, color='g',label='Versicolor')
axs[2, 1].hist(virginica_petal_length, bins=10, alpha=0.7, rwidth=0.85, color='b',label='Virginica')
axs[2, 1].set_title('Petal Length Distribution across Species')
axs[2, 1].set_xlabel('Petal Length')
axs[2, 1].set_ylabel('Frequency')
axs[2, 1].legend()

plt.tight_layout()  # prevent overlap
plt.show()  # render plots
```

**Between Species**

*Setosa*

- *Short* but *wide* sepals 
- *Short* and *narrow* petals
- Based on the graph, its *short* and *wide* sepals combined with its *short* petals make it *easily* distinguishable from the other species.

## 5. Create more charts (different than for the class comparison, use matplotlib)

```python
# Calculate Areas
setosa_area_sepal = setosa_sepal_width*setosa_sepal_length 
versicolor_area_sepal = versicolor_sepal_width * versicolor_sepal_length
virginica_area_sepal = virginica_sepal_width * virginica_sepal_length 

setosa_area_petal = setosa_petal_width*setosa_petal_length
versicolor_area_petal = versicolor_petal_width * versicolor_petal_length 
virginica_area_petal = virginica_petal_width * virginica_petal_length 

# Calculate Ratios
setosa_petal_to_sepal_length_ratio = setosa_petal_length / setosa_sepal_length
versicolor_petal_to_sepal_length_ratio = versicolor_petal_length / versicolor_sepal_length
virginica_petal_to_sepal_length_ratio = virginica_petal_length / virginica_sepal_length

setosa_petal_to_sepal_width_ratio = setosa_petal_width / setosa_sepal_width
versicolor_petal_to_sepal_width_ratio = versicolor_petal_width / versicolor_sepal_width
virginica_petal_to_sepal_width_ratio = virginica_petal_width / virginica_sepal_width

# 2x2 subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(14, 12))

# (1) Sepal Area
axs[0,0].hist([setosa_area_sepal, versicolor_area_sepal, virginica_area_sepal], label=['setosa', 'versicolor', 'virginica'], color=['red','blue','orange'],bins=10, alpha=0.7)

setosa_area_sepal_mean = np.mean(setosa_area_sepal)
versicolor_area_sepal_mean = np.mean(versicolor_area_sepal)
virginica_area_sepal_mean = np.mean(virginica_area_sepal)

setosa_area_sepal_stdv = np.std(setosa_area_sepal)
versicolor_area_sepal_stdv = np.std(versicolor_area_sepal)
virginica_area_sepal_stdv = np.std(virginica_area_sepal)

# Upper standard deviation range
axs[0,0].axvline(setosa_area_sepal_mean + setosa_area_sepal_stdv, color='red', linestyle='-', linewidth=0.95)
axs[0,0].axvline(versicolor_area_sepal_mean + versicolor_area_sepal_stdv, color='blue', linestyle='-', linewidth=0.95)
axs[0,0].axvline(virginica_area_sepal_mean + virginica_area_sepal_stdv, color='orange', linestyle='-', linewidth=0.95)

# Lower standard deviation range
axs[0,0].axvline(setosa_area_sepal_mean - setosa_area_sepal_stdv, color='red', linestyle='-', linewidth=0.95)
axs[0,0].axvline(versicolor_area_sepal_mean - versicolor_area_sepal_stdv, color='blue', linestyle='-', linewidth=0.95)
axs[0,0].axvline(virginica_area_sepal_mean - virginica_area_sepal_stdv, color='orange', linestyle='-', linewidth=0.95)

axs[0,0].set_title('Sepal Area')
axs[0,0].legend()

# (2) Petal Area
axs[0,1].hist([setosa_area_petal, versicolor_area_petal, virginica_area_petal],  color=['red','blue','orange'], label=['Setosa', 'Versicolor', 'Virginica'], bins=10, alpha=0.7)
axs[0,1].set_title('Petal Area')
axs[0,1].legend()

# (3) Petal to Sepal Length Ratio
axs[1,0].hist([setosa_petal_to_sepal_length_ratio, versicolor_petal_to_sepal_length_ratio, virginica_petal_to_sepal_length_ratio], color=['red','blue','orange'], label=['Setosa', 'Versicolor', 'Virginica'], bins=10, alpha=0.7)
axs[1,0].set_title('Petal to Sepal Length Ratio')
axs[1,0].legend()

# (4) Petal to Sepal Width Ratio
axs[1,1].hist([setosa_petal_to_sepal_width_ratio, versicolor_petal_to_sepal_width_ratio, virginica_petal_to_sepal_width_ratio], color=['red','blue','orange'],label=['Setosa', 'Versicolor', 'Virginica'], bins=10, alpha=0.7)
axs[1,1].set_title('Petal to Sepal Width Ratio')
axs[1,1].legend()

plt.tight_layout()
plt.show()
```

## 1. Sepal Area
    
- Given that the majority of data points cluster within a standard deviation and that the peak of Setosa closely matches its lower bound, it appears that most Sectosa flowers have sepal areas that are closer to the lower end of their standard deviation range. The common sepal area of the Versicolor species is located in the middle of its standard deviation range, indicating that the majority of these flowers have medium-sized sepal areas. In general, Virginia species have bigger sepal areas, which are more in line with their higher standard deviation. The range of sepal areas for Versicolor and Setosa overlaps, making it challenging to distinguish between the two species only in sepal areas.
    
## 2. Petal Area
    
- When you look at the petal areas, you can tell the three species apart. The smallest petal area is on Setosa flowers, the middle one is on Versicolor flowers, and the biggest is on Virginica flowers. This metric can help you tell the difference between the three species, even between the Versicolor and Virginica species, which are often confused with each other.


## 3. Petal to Sepal Length Ratio

- Setosa's petal-to-sepal length ratio is shorter than those of the other two species, indicating that its petals are proportionally shorter than its sepals. Both Versicolor and Virginica present higher ratios. However, Virginica tends to have a slightly higher ratio than Versicolor, indicating that its petals are slightly longer in relation to its sepals. While this metric is useful for distinguishing Setosa, it may not be the most reliable for separating Versicolor and Virginica.


## 4. Petal to Sepal Width Ratio
    
- Setosa displays a lower petal-to-sepal width ratio, indicating that its petals are narrower than its sepals. Both Versicolor and Virginica have higher ratios, but Virginica has a slightly wider ratio, indicating that its petals are wider in comparison to its sepals. This trait effectively distinguishes Setosa, but it may not be the best criteria for differentiating Versicolor and Virginica.
